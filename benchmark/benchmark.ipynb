```python
import pandas as pd
import numpy as np
import bokeh.plotting as bp
import bokeh.palettes as bpal
bp.output_notebook()
```

```python
benchmark_results = pd.read_csv('../cmake-build-minsizerel/benchmark/bm_TriangulationFlipGraph.csv', skiprows=8)
benchmark_results["suite_name"] = benchmark_results["name"].apply(lambda x: x.split('/')[0]).str.removesuffix("/real_time").str.removesuffix("/iterations:10").str.removeprefix("BM_").str.replace(", PointSet1", "")
benchmark_results["n_points"] = benchmark_results["name"].apply(lambda x: x.split('/')[1]).astype(int)
benchmark_results
```

```python
pl = bp.figure(x_axis_label='Number of points', y_axis_label='Real time (s)', y_axis_type="log", width=800, height=600)

for (suite_name, dfg), c in zip(benchmark_results.groupby('suite_name'), bpal.Category10[10]):
    pl.line(dfg['n_points'], dfg['real_time'], legend_label=suite_name, color=c)
    pl.scatter(dfg['n_points'], dfg['real_time'], legend_label=suite_name, color=c)

pl.legend.location = "top_left"
    
bp.show(pl)





pl2 = bp.figure(x_axis_label='Number of points', y_axis_label='Parallel efficiency (%)', width=800, height=600)

for (suite_name, dfg), c in zip(benchmark_results.groupby('suite_name'), bpal.Category10[10]):
    pl2.line(dfg['n_points'], 100*dfg['cpu_time']/dfg['real_time'], legend_label=suite_name, color=c)
    pl2.scatter(dfg['n_points'], 100*dfg['cpu_time']/dfg['real_time'], legend_label=suite_name, color=c)

pl2.legend.location = "bottom_left"
    
bp.show(pl2)
```
